```python
from sklearn.linear_model import Ridge
import lightgbm
from lightgbm import LGBMRegressor
import xgboost
from xgboost import XGBRegressor
from sklearn.ensemble import ExtraTreesRegressor
import catboost
from catboost import CatBoostRegressor
```

```python
validation =0
x = 1
na_val=-999
```

```python
if validation==1:
    #df = data[data['Time']!='Feb-2023']
    df = history[history["Month"]!='28-02-2023'].iloc[:,:3]
else:
    #df=data.copy()
    df=history.copy()
sub=pd.DataFrame()
check=pd.DataFrame()
df = df.rename({"Month":"Date"},axis=1)
CategoryEncoder = LabelEncoder()
df["Cat"] = CategoryEncoder.fit_transform(df["Category"]).astype("int")
df[["Day","Month","Year"]] = df["Date"].str.split("-",expand=True).astype("int")
df["Year"] = df["Year"] - df["Year"].min()
df["Timestamp"] = (df["Year"]*12)+df["Month"]
df = df.sort_values(["Category","Timestamp"])
for i in range(1,13,1):
    print(i)
    df[f"last{i}"] = df.groupby(["Category"])["Value"].shift(i)
    df[f"last{i}_"] = df[f"last{i}"]/df["Value"]
#df["mean_t"] = df[["last1","Value"]].mean(axis=1)
df["target"] = df.groupby(["Category"])["Value"].shift(-1*x)
#df=pd.get_dummies(df,columns=["Cat"])
target = "target"
drop_cols=["Date","Category",target]
if validation==1:
    tdf=df[df[target].isna()]
else:
    tdf=df[df["Timestamp"]==df["Timestamp"].max()]
df=df[df[target].notna()]
cols = df.drop(drop_cols,axis=1).columns
if validation==1:
    tdf=tdf[tdf["Timestamp"]==tdf["Timestamp"].max()-(x-1)]
else:
    tdf = tdf[tdf["Timestamp"]==tdf["Timestamp"].max()]
model = Ridge()
#model = CatBoostRegressor(silent=True,cat_features=["Cat"])
#model = ExtraTreesRegressor(random_state=0,n_estimators=20)
#model = LGBMRegressor(n_estimators=100)#,objective="tweedie")
model.fit(df[cols].fillna(na_val),df[target])
tdf[target]= model.predict(tdf[cols].fillna(na_val))
if validation==1:
    check = pd.merge(scorer,tdf[["Category",target]],on="Category",how="left")
    print(mse(check["Value"],check[target])**0.5)
else:
    tdf["ID"]="April_"+tdf.Category.astype(str)
    tdf["Value"] = tdf[target].values
    sub=tdf[tdf.ID.isin(ss.ID)][ss.columns]
    sub.to_csv("sub.csv",index=None)
df
```
